```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the Plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, label = "Boston", color = "green", marker = "D", linewidth=2)
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_xlim(0,11)
ax.set_ylim(0,45)
ax.grid()
ax.set_title("PyBer Fare by Month")
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust the size by multiplying the data in the y-axis by any value. 
# multiply each data point in the y-axis by 3 
#iterate through the y-axis data and multiply each data point by 3 and add it to a new list
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#  list comprehension technique to multiply each data point in the y-axis list
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="blue", edgecolor="black")
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    shadow=bool,
    labels=x_axis,
    autopct='%.1f%%')
plt.gca().invert_yaxis()
plt.show()
```
